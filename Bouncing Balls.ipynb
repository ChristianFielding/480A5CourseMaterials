```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import random
import numpy as np

# Create a figure and a set of subplots
fig, ax = plt.subplots()

# Number of balls
num_balls = 50

# Initial positions and velocities of the balls
positions = np.random.rand(num_balls, 2) * 10
velocities = np.random.randn(num_balls, 2) * 0.1

# Set the limits and labels
ax.set_xlim([0, 10])
ax.set_ylim([0, 10])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('2D Box')

time_steps = 100
gravity = -0.01

# Function to update the plot
def update(frame):
    global positions, velocities
    
    # Update positions with velocities
    positions += velocities
    
    # Apply gravity
    velocities[:, 1] += gravity
    
    # Check for boundaries and bounce if necessary
    for i in range(num_balls):
        if positions[i, 0] < 0 or positions[i, 0] > 10:
            velocities[i, 0] = -velocities[i, 0]
        if positions[i, 1] < 0 or positions[i, 1] > 10:
            velocities[i, 1] = -velocities[i, 1]
    
    # Check for collisions between balls
    for i in range(num_balls):
        for j in range(i + 1, num_balls):
            dist = np.linalg.norm(positions[i] - positions[j])
            if dist < 0.2:  # Assuming each ball has a radius of 0.1
                velocities[i], velocities[j] = velocities[j], velocities[i]
    
    # Clear the previous points
    ax.cla()
    
    # Plot the new points
    ax.scatter(positions[:, 0], positions[:, 1])
    
    # Reset the limits and labels
    ax.set_xlim([0, 10])
    ax.set_ylim([0, 10])
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('2D Box')

# Create the animation
ani = animation.FuncAnimation(fig, update, frames=time_steps, repeat=False)

# Save the animation as a video file
ani.save('brownian_motion_with_collisions.gif', writer='pillow', fps=10)

plt.show()
```
